```python
# imports
import pandas as pd
from keras import models
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

from tensorflow.keras.constraints import MaxNorm
from tensorflow.keras.layers import Dropout
```

```python
# load dataset
df = pd.read_csv('/Users/taylor/Desktop/DS340W/term_project/data/wisconsin/wisconsin_train_balanced.csv')
dataset = df.values

# split into input (X) and output (Y) variables
X = dataset[:,0:30].astype(float)
Y = dataset[:,30]

# encode class values as integers
encoder = LabelEncoder()

encoder.fit(Y)
encoded_Y = encoder.transform(Y)
```

```python
# Create 4 larger models (i.e., 5, 7, 10, 15 hidden layers):
```

```python
### 5-hidden layers #####
bigNet5Drop = Sequential()

bigNet5Drop.add(Dense(30, input_shape=(30,), activation='relu', kernel_constraint=MaxNorm(3)))
bigNet5Drop.add(Dropout(0.2))
bigNet5Drop.add(Dense(15, activation='relu', kernel_constraint=MaxNorm(3)))
bigNet5Drop.add(Dropout(0.2))
bigNet5Drop.add(Dense(15, activation='relu', kernel_constraint=MaxNorm(3)))
bigNet5Drop.add(Dropout(0.2))
bigNet5Drop.add(Dense(15, activation='relu', kernel_constraint=MaxNorm(3)))
bigNet5Drop.add(Dropout(0.2))
bigNet5Drop.add(Dense(15, activation='relu', kernel_constraint=MaxNorm(3)))
bigNet5Drop.add(Dropout(0.2))

bigNet5Drop.add(Dense(1, activation='sigmoid'))

# Compile and train
bigNet5Drop.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
bigNet5Drop.fit(X, encoded_Y, batch_size = 32, epochs = 100)
```
