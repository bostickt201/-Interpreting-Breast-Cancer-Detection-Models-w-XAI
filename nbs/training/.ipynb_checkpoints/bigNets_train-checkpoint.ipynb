```python
# imports
import pandas as pd
from keras import models
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
# load dataset
df = pd.read_csv('/Users/taylor/Desktop/DS340W/term_project/data/wisconsin/wisconsin_train_balanced.csv')
dataset = df.values

# split into input (X) and output (Y) variables
X = dataset[:,0:30].astype(float)
Y = dataset[:,30]

# encode class values as integers
encoder = LabelEncoder()

encoder.fit(Y)
encoded_Y = encoder.transform(Y)
```

```python
# Create 4 larger models (i.e., 5, 7, 10, 15 hidden layers):
```

```python
### 5-hidden layers #####
bigNet5 = Sequential()
bigNet5.add(Dense(30, input_shape=(30,), activation='relu'))
bigNet5.add(Dense(15, activation='relu'))
bigNet5.add(Dense(15, activation='relu'))
bigNet5.add(Dense(15, activation='relu'))
bigNet5.add(Dense(15, activation='relu'))
bigNet5.add(Dense(1, activation='sigmoid'))

# Compile and train
bigNet5.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
bigNet5.fit(X, encoded_Y, batch_size = 32, epochs = 100)
```

```python
### 7-hidden layers #####
bigNet7 = Sequential()
bigNet7.add(Dense(30, input_shape=(30,), activation='relu'))
bigNet7.add(Dense(15, activation='relu'))
bigNet7.add(Dense(15, activation='relu'))
bigNet7.add(Dense(15, activation='relu'))
bigNet7.add(Dense(15, activation='relu'))
bigNet7.add(Dense(15, activation='relu'))
bigNet7.add(Dense(15, activation='relu'))
bigNet7.add(Dense(1, activation='sigmoid'))

# Compile and train
bigNet7.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
bigNet7.fit(X, encoded_Y, batch_size = 64, epochs = 100) # increased batch size from 32 to 64
```

```python
### 10-hidden layers #####
bigNet10 = Sequential()
bigNet10.add(Dense(30, input_shape=(30,), activation='relu'))
bigNet10.add(Dense(15, activation='relu'))
bigNet10.add(Dense(15, activation='relu'))
bigNet10.add(Dense(15, activation='relu'))
bigNet10.add(Dense(15, activation='relu'))
bigNet10.add(Dense(15, activation='relu'))
bigNet10.add(Dense(15, activation='relu'))
bigNet10.add(Dense(15, activation='relu'))
bigNet10.add(Dense(15, activation='relu'))
bigNet10.add(Dense(15, activation='relu'))
bigNet10.add(Dense(1, activation='sigmoid'))

# Compile and train
bigNet10.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
bigNet10.fit(X, encoded_Y, batch_size = 32, epochs = 100)
```

```python
### 15-hidden layers #####
bigNet15 = Sequential()
bigNet15.add(Dense(30, input_shape=(30,), activation='relu'))
bigNet15.add(Dense(15, activation='relu'))
bigNet15.add(Dense(15, activation='relu'))
bigNet15.add(Dense(15, activation='relu'))
bigNet15.add(Dense(15, activation='relu'))
bigNet15.add(Dense(15, activation='relu'))
bigNet15.add(Dense(15, activation='relu'))
bigNet15.add(Dense(15, activation='relu'))
bigNet15.add(Dense(15, activation='relu'))
bigNet15.add(Dense(15, activation='relu'))
bigNet15.add(Dense(15, activation='relu'))
bigNet15.add(Dense(15, activation='relu'))
bigNet15.add(Dense(15, activation='relu'))
bigNet15.add(Dense(15, activation='relu'))
bigNet15.add(Dense(15, activation='relu'))
bigNet15.add(Dense(1, activation='sigmoid'))

# Compile and train
bigNet15.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
bigNet15.fit(X, encoded_Y, batch_size = 64, epochs = 100) # increased batch size from 32 to 64
```
